```python
path = "models/AllenCahnSingle0/PINNfull0"

subdirs = []
for entry_name in os.listdir(path):
        full_path = os.path.join(path, entry_name)
        if os.path.isdir(full_path):
                subdirs.append(entry_name)

selection = [
        "trial0",
        "trial1",
        "trial2",
        "trial3",
        "trial4",
        "trial5",
        "trial6",
        "trial7",
        "trial8",
        "trial9"
        ]

labels = [trial for trial in os.listdir(path) if trial in selection]
filepaths=[os.path.join(path, trial) + "/train_stats.npy" for trial in subdirs if trial in selection]

plot_model_stats(filepaths=filepaths, labels=labels, to_plot=["train"])

filepaths=[os.path.join(path, trial) + "/test_stats.npy" for trial in subdirs if trial in selection]

plot_model_stats(filepaths=filepaths, labels=labels, to_plot=["test"])
```

```python
best_trial = "trial5"
labels = [best_trial, best_trial]
filepaths=[os.path.join(path, best_trial) + "/train_stats.npy", os.path.join(path, best_trial) + "/test_stats.npy"]
plot_model_stats(filepaths=filepaths, labels=labels, to_plot=["train", "test"])
print_model_info(filepath=os.path.join(path, best_trial) + "/model.pth")
```
